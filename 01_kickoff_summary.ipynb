
Goal:
To predict the final cost of an insurance claim as early as possible, using FNOL-stage information, to improve reserve setting and claims handling.

Target: 
Ultimate Claim Amount
This is a regression problem because we are predicting continous monetary value

Reason:
Financial control: Better early estimates reduce over- and under-reserving, improving capital efficiency.
Operational efficiency: High-risk, high-cost claims can be flagged early for priority handling.
Customer outcomes: Faster, more confident decisions reduce delays and disputes.

Definitions
FNOL (First Notification of Loss):
This is the point at which the customer first reports an accident or loss to GMA. Information at this stage is partial and uncertain, but it is used to open the claim and set initial financial reserves. 

Estimated Claim Amount:
This refers to early estimate of the expected cost of the claim, calculated at or shortly after FNOL. It is based on limited details such as claim type, driver profile, vehicle details, and initial accident information. This is what is used to set reserves but can be inaccurate due to missing information. 

Ultimate Claim Amount:
This is the final total payout for the claim.
This reflects all costs incurred, including repairs, injuries, third-party costs, and legal expenses. It is the target variable the machine learning model aims to predict. 

Third-party severity:
A measure of how serious the impact of an accident is on third parties.
In the dataset, this is captured through indicators such as: Number of third parties involved, Injury severity categories (minor, serious, fatal)

Modelling Metrics:
Because this is a regression problem with a monetary target, the metrics must measure error size not correctness
Primary Metrics 
•	MAE (Mean Absolute Error): This will be used first to measure absolute error in predictions. This is important because insurance cares about average reserving error. 
•	RMSE (Root Mean Square Error): This highlights failures in high claims and its useful for risk awareness
•	MAPE (Mean Absolute Percentage Error): Optional
Measures error relative to actual claim size
